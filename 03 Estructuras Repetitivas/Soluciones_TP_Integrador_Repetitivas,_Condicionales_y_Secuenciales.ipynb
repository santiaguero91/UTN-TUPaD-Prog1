# Ejercicio 1 — “Caja del Kiosco” 🧾

## 🎯 Objetivo
Simular una compra con validaciones y cálculo de totales.

---

## 📌 Consigna
1. Pedir **nombre del cliente**  
   - Debe contener **solo letras**  
   - Validar con `.isalpha()` dentro de un `while`  
   - **No aceptar vacío** (si está vacío, es error)

2. Pedir **cantidad de productos** a comprar  
   - Debe ser un **entero positivo**  
   - Validar con `.isdigit()` dentro de un `while`  
   - Debe ser **> 0** (si ingresa 0, volver a pedir)

3. Por cada producto (usar `for`):
   - Pedir **precio** (entero)  
     - Validar con `.isdigit()` (sin `try/except`)
   - Pedir si tiene **descuento S/N**  
     - Validar con `while`
     - Aceptar `s` o `n` en cualquier mayúscula/minúscula (`S`, `s`, `N`, `n`)
   - Si tiene descuento, aplicar **10%** al precio de ese producto

4. Al final mostrar:
   - **Total sin descuentos**
   - **Total con descuentos**
   - **Ahorro total**
   - **Promedio por producto** (usar `float` y formatear con `:.2f`)

---

## ✅ Validaciones obligatorias
- **Sin `try/except`.**
- **No aceptar vacío** en el nombre.
- **Cantidad > 0.**

---

## 🧪 Salida esperada (ejemplo)
Cliente: Ana  
Cantidad de productos: 3  
Producto 1 - Precio: 100 Descuento (S/N): s  
Producto 2 - Precio: 50 Descuento (S/N): n  
Producto 3 - Precio: 200 Descuento (S/N): s  

Total sin descuentos: $350  

Total con descuentos: $320.00  

Ahorro: $30.00  

Promedio por producto: $106.67


```python
# ---------------------------
# 1) PEDIR Y VALIDAR NOMBRE
# ---------------------------
nombre = input("Cliente: ").strip()  # strip() elimina espacios al inicio/fin

# Repetimos mientras:
# - esté vacío (no se acepta)
# - o tenga caracteres que no sean letras (isalpha() valida solo letras, sin espacios)
while nombre == "" or not nombre.isalpha():
    print("Error: ingresá un nombre no vacío y solo con letras (sin espacios ni números).")
    nombre = input("Cliente: ").strip()

# -----------------------------------------
# 2) PEDIR Y VALIDAR CANTIDAD DE PRODUCTOS
# -----------------------------------------
cantidad_str = input("Cantidad de productos: ").strip()

# Validación con while:
# - isdigit() asegura que sean solo dígitos (ej: "3")
# - además, debe ser > 0 (si ingresa "0", se vuelve a pedir)
while not cantidad_str.isdigit() or int(cantidad_str) <= 0:
    print("Error: ingresá un número entero positivo mayor que 0.")
    cantidad_str = input("Cantidad de productos: ").strip()

# Convertimos a entero cuando ya está validado
cantidad = int(cantidad_str)

# -----------------------------------------
# 3) CARGA DE PRODUCTOS + DESCUENTOS
# -----------------------------------------
total_sin_descuento = 0          # suma de precios originales
total_con_descuento = 0.0        # suma aplicando descuentos (float por el 10%)

for i in range(1, cantidad + 1):
    # ---- Pedir y validar precio ----
    precio_str = input(f"Producto {i} - Precio: ").strip()

    # Precio debe ser entero positivo (digit) -> por consigna "entero"
    # (si quisieras permitir 0, se cambia la condición, pero usualmente no tiene sentido)
    while not precio_str.isdigit() or int(precio_str) <= 0:
        print("Error: ingresá un precio entero positivo (solo números, mayor que 0).")
        precio_str = input(f"Producto {i} - Precio: ").strip()

    precio = int(precio_str)

    # ---- Pedir y validar descuento S/N ----
    desc = input("Descuento (S/N): ").strip().lower()
    # Aceptamos únicamente "s" o "n"
    while desc not in ("s", "n"):
        print("Error: ingresá 'S' para sí o 'N' para no (se acepta mayúscula o minúscula).")
        desc = input("Descuento (S/N): ").strip().lower()

    # ---- Acumular totales ----
    total_sin_descuento += precio

    # Si hay descuento: 10% de descuento => pagar 90% del precio
    if desc == "s":
        precio_final = precio * 0.90
    else:
        precio_final = float(precio)

    total_con_descuento += precio_final

# -----------------------------------------
# 4) RESULTADOS FINALES
# -----------------------------------------
ahorro = total_sin_descuento - total_con_descuento
promedio = total_con_descuento / cantidad  # promedio por producto (float)

print()  # línea en blanco para prolijidad
print(f"Total sin descuentos: ${total_sin_descuento}")
print(f"Total con descuentos: ${total_con_descuento:.2f}")
print(f"Ahorro: ${ahorro:.2f}")
print(f"Promedio por producto: ${promedio:.2f}")
```

# Ejercicio 2 — “Acceso al Campus y Menú Seguro” 🔐

## 🎯 Objetivo
Simular un login con intentos limitados y, si el acceso es correcto, mostrar un menú seguro con validaciones estrictas.

---

## 📌 Requisitos

### 1) Credenciales fijas
Definir en el código:
- **Usuario correcto:** `alumno`
- **Clave correcta:** `python123`

### 2) Intentos de login
- Permitir **máximo 3 intentos** para ingresar usuario y clave.
- Si falla 3 veces: mostrar **"Cuenta bloqueada"** y terminar el programa.
- Si acierta: mostrar **"Acceso concedido."**

### 3) Menú repetitivo (while)
Una vez dentro, mostrar un menú que se repite hasta elegir salir:

1. Ver estado de inscripción (mostrar **"Inscripto"**)
2. Cambiar clave  
   - Pedir **nueva clave** y **confirmación**
   - Deben **coincidir**
   - La clave debe tener **mínimo 6 caracteres** (`len()`)
3. Mostrar mensaje motivacional (1 frase)
4. Salir

### 4) Validación del menú (obligatoria)
- Debe ser número: `.isdigit()`
- Debe estar en el rango **1 a 4**

---

## 🧪 Salida esperada (ejemplo)
Intento 1/3 - Usuario: alumno  
Clave: xxx  
Error: credenciales inválidas.  

Intento 2/3 - Usuario: alumno  
Clave: python123  
Acceso concedido.  

1) Estado 2) Cambiar clave 3) Mensaje 4) Salir  
Opción: a  
Error: ingrese un número válido.  
Opción: 5  
Error: opción fuera de rango.  
Opción: 2  
Nueva clave: 123  
Error: mínimo 6 caracteres.  
...


```python
# Ejercicio 2 — “Acceso al Campus y Menú Seguro”

USUARIO_CORRECTO = "alumno"
CLAVE_CORRECTA = "python123"

# La clave actual puede cambiar si el usuario elige "Cambiar clave"
clave_actual = CLAVE_CORRECTA

# ----------------------------
# LOGIN CON MÁXIMO 3 INTENTOS
# ----------------------------
intento = 1
acceso_concedido = False

while intento <= 3 and acceso_concedido == False:
    usuario = input(f"Intento {intento}/3 - Usuario: ").strip()
    clave = input("Clave: ").strip()

    if usuario == USUARIO_CORRECTO and clave == clave_actual:
        print("Acceso concedido.")
        acceso_concedido = True
    else:
        print("Error: credenciales inválidas.")
        intento += 1

# Si no accedió en 3 intentos, se bloquea
if acceso_concedido == False:
    print("Cuenta bloqueada")
else:
    # ----------------
    # MENÚ REPETITIVO
    # ----------------
    seguir_menu = True

    while seguir_menu == True:
        print("1) Estado 2) Cambiar clave 3) Mensaje 4) Salir")
        opcion_str = input("Opción: ").strip()

        # ---- Validación de "número" ----
        while not opcion_str.isdigit():
            print("Error: ingrese un número válido.")
            opcion_str = input("Opción: ").strip()

        opcion = int(opcion_str)

        # ---- Validación de rango 1..4 ----
        while opcion < 1 or opcion > 4:
            print("Error: opción fuera de rango.")
            opcion_str = input("Opción: ").strip()

            while not opcion_str.isdigit():
                print("Error: ingrese un número válido.")
                opcion_str = input("Opción: ").strip()

            opcion = int(opcion_str)

        # ----------------------------
        # ACCIONES DEL MENÚ
        # ----------------------------
        if opcion == 1:
            print("Inscripto")

        elif opcion == 2:
            # Cambiar clave: usamos una bandera de "cambio_exitoso"
            cambio_exitoso = False

            while cambio_exitoso == False:
                nueva = input("Nueva clave: ").strip()

                # Validación mínimo 6 caracteres
                if len(nueva) < 6:
                    print("Error: mínimo 6 caracteres.")
                else:
                    confirmacion = input("Confirmar clave: ").strip()

                    if nueva != confirmacion:
                        print("Error: las claves no coinciden.")
                    else:
                        clave_actual = nueva
                        print("Clave actualizada correctamente.")
                        cambio_exitoso = True

        elif opcion == 3:
            print("¡Paso a paso: si insistís, te sale!")

        elif opcion == 4:
            print("Sesión cerrada.")
            seguir_menu = False
```

# Ejercicio 3 (Alta) — “Agenda de Turnos con Nombres (sin listas)” 🗓️

## 🌟 Contexto
Hay 2 días de atención:
- **Lunes**: 4 turnos
- **Martes**: 3 turnos

---

## 📌 Reglas

### 1) Operador
Pedir **nombre del operador** (solo letras).

### 2) Menú repetitivo (hasta cerrar sistema)
1. Reservar turno
2. Cancelar turno (por nombre)
3. Ver agenda del día
4. Ver resumen general
5. Cerrar sistema

### 3) Reservar turno
- Elegir día: **1 = Lunes**, **2 = Martes**
- Pedir **nombre del paciente** (solo letras)
- Verificar que el paciente **no esté repetido en ese día** (comparando con variables)
- Guardar en el **primer espacio libre** del día (`lunes1`, `lunes2`, ...)

### 4) Cancelar turno
- Elegir día
- Pedir nombre del paciente (solo letras)
- Si existe en ese día, cancelar y dejar el espacio vacío: `""`

### 5) Ver agenda del día
Mostrar los turnos en orden (`Turno 1..N`), indicando **"(libre)"** si está vacío.

### 6) Resumen general
- Turnos **ocupados** y **disponibles** por día
- Día con **más turnos** ocupados (o **empate**)

---

## ⛔ Restricciones
- ❌ No usar **listas**, **diccionarios**, **sets**, **tuplas**
- ✅ Se permite usar `""` como “vacío”
- ✅ Validaciones con `.isalpha()` y `.isdigit()` (sin `try/except`)


```python
# ----------------------------------------
# 0) TURNOS DISPONIBLES (variables sueltas)
# ----------------------------------------
# Lunes tiene 4 turnos
lunes1 = ""
lunes2 = ""
lunes3 = ""
lunes4 = ""

# Martes tiene 3 turnos
martes1 = ""
martes2 = ""
martes3 = ""

# ----------------------------------------
# 1) PEDIR NOMBRE DEL OPERADOR (solo letras)
# ----------------------------------------
operador = input("Nombre del operador: ").strip()

# Validación: no vacío y solo letras
while operador == "" or not operador.isalpha():
    print("Error: el nombre del operador debe ser no vacío y solo con letras.")
    operador = input("Nombre del operador: ").strip()

print(f"\nSistema iniciado. Operador: {operador}\n")

# ----------------------------------------
# 2) MENÚ REPETITIVO HASTA CERRAR (opción 5)
# ----------------------------------------
opcion = 0  # arrancamos distinto de 5 para entrar al bucle

while opcion != 5:
    # Mostramos el menú
    print("1) Reservar turno")
    print("2) Cancelar turno (por nombre)")
    print("3) Ver agenda del día")
    print("4) Ver resumen general")
    print("5) Cerrar sistema")

    opcion_str = input("Opción: ").strip()

    # Validación: debe ser un número del 1 al 5
    while (not opcion_str.isdigit()) or (int(opcion_str) < 1) or (int(opcion_str) > 5):
        if not opcion_str.isdigit():
            print("Error: ingrese un número válido.")
        else:
            print("Error: opción fuera de rango (1 a 5).")
        opcion_str = input("Opción: ").strip()

    opcion = int(opcion_str)

    # ----------------------------------------
    # OPCIÓN 1: RESERVAR TURNO
    # ----------------------------------------
    if opcion == 1:
        # Elegir día (1=Lunes, 2=Martes)
        dia_str = input("Día (1=Lunes, 2=Martes): ").strip()

        # Validación estricta del día (solo 1 o 2)
        while (not dia_str.isdigit()) or (int(dia_str) < 1) or (int(dia_str) > 2):
            if not dia_str.isdigit():
                print("Error: ingrese un número válido.")
            else:
                print("Error: opción fuera de rango (1 o 2).")
            dia_str = input("Día (1=Lunes, 2=Martes): ").strip()

        dia = int(dia_str)

        # Pedir nombre del paciente (solo letras)
        paciente = input("Nombre del paciente: ").strip()

        while paciente == "" or not paciente.isalpha():
            print("Error: el nombre del paciente debe ser no vacío y solo con letras.")
            paciente = input("Nombre del paciente: ").strip()

        # ----------------------------------------
        # Verificar repetido en el día elegido
        # ----------------------------------------
        repetido = False  # bandera: asumimos que NO está repetido hasta verificar

        if dia == 1:
            # Comparación contra todos los turnos del lunes
            if paciente == lunes1 or paciente == lunes2 or paciente == lunes3 or paciente == lunes4:
                repetido = True
        else:
            # dia == 2 -> Comparación contra todos los turnos del martes
            if paciente == martes1 or paciente == martes2 or paciente == martes3:
                repetido = True

        if repetido == True:
            print("Error: ese paciente ya tiene un turno reservado en ese día.\n")
        else:
            # ----------------------------------------
            # Guardar en el primer espacio libre
            # ----------------------------------------
            reservado = False  # bandera para saber si encontramos un lugar

            if dia == 1:
                # Buscar el primer turno libre del lunes
                if lunes1 == "":
                    lunes1 = paciente
                    reservado = True
                    print("Turno reservado en Lunes - Turno 1.\n")
                elif lunes2 == "":
                    lunes2 = paciente
                    reservado = True
                    print("Turno reservado en Lunes - Turno 2.\n")
                elif lunes3 == "":
                    lunes3 = paciente
                    reservado = True
                    print("Turno reservado en Lunes - Turno 3.\n")
                elif lunes4 == "":
                    lunes4 = paciente
                    reservado = True
                    print("Turno reservado en Lunes - Turno 4.\n")
            else:
                # Buscar el primer turno libre del martes
                if martes1 == "":
                    martes1 = paciente
                    reservado = True
                    print("Turno reservado en Martes - Turno 1.\n")
                elif martes2 == "":
                    martes2 = paciente
                    reservado = True
                    print("Turno reservado en Martes - Turno 2.\n")
                elif martes3 == "":
                    martes3 = paciente
                    reservado = True
                    print("Turno reservado en Martes - Turno 3.\n")

            # Si no se reservó, es porque el día estaba lleno
            if reservado == False:
                print("No hay turnos disponibles en ese día.\n")

    # ----------------------------------------
    # OPCIÓN 2: CANCELAR TURNO (por nombre)
    # ----------------------------------------
    elif opcion == 2:
        dia_str = input("Día (1=Lunes, 2=Martes): ").strip()

        while (not dia_str.isdigit()) or (int(dia_str) < 1) or (int(dia_str) > 2):
            if not dia_str.isdigit():
                print("Error: ingrese un número válido.")
            else:
                print("Error: opción fuera de rango (1 o 2).")
            dia_str = input("Día (1=Lunes, 2=Martes): ").strip()

        dia = int(dia_str)

        paciente = input("Nombre del paciente a cancelar: ").strip()
        while paciente == "" or not paciente.isalpha():
            print("Error: el nombre del paciente debe ser no vacío y solo con letras.")
            paciente = input("Nombre del paciente a cancelar: ").strip()

        # Buscamos si existe y lo vaciamos (""), sin listas: comparando variable por variable
        encontrado = False

        if dia == 1:
            if lunes1 == paciente:
                lunes1 = ""
                encontrado = True
            elif lunes2 == paciente:
                lunes2 = ""
                encontrado = True
            elif lunes3 == paciente:
                lunes3 = ""
                encontrado = True
            elif lunes4 == paciente:
                lunes4 = ""
                encontrado = True
        else:
            if martes1 == paciente:
                martes1 = ""
                encontrado = True
            elif martes2 == paciente:
                martes2 = ""
                encontrado = True
            elif martes3 == paciente:
                martes3 = ""
                encontrado = True

        if encontrado == True:
            print("Turno cancelado correctamente.\n")
        else:
            print("No se encontró un turno con ese nombre en ese día.\n")

    # ----------------------------------------
    # OPCIÓN 3: VER AGENDA DEL DÍA
    # ----------------------------------------
    elif opcion == 3:
        dia_str = input("Día (1=Lunes, 2=Martes): ").strip()

        while (not dia_str.isdigit()) or (int(dia_str) < 1) or (int(dia_str) > 2):
            if not dia_str.isdigit():
                print("Error: ingrese un número válido.")
            else:
                print("Error: opción fuera de rango (1 o 2).")
            dia_str = input("Día (1=Lunes, 2=Martes): ").strip()

        dia = int(dia_str)

        if dia == 1:
            print("\nAgenda del Lunes:")
            # Turno 1
            if lunes1 == "":
                print("Turno 1: (libre)")
            else:
                print(f"Turno 1: {lunes1}")
            # Turno 2
            if lunes2 == "":
                print("Turno 2: (libre)")
            else:
                print(f"Turno 2: {lunes2}")
            # Turno 3
            if lunes3 == "":
                print("Turno 3: (libre)")
            else:
                print(f"Turno 3: {lunes3}")
            # Turno 4
            if lunes4 == "":
                print("Turno 4: (libre)")
            else:
                print(f"Turno 4: {lunes4}")
            print()  # línea en blanco
        else:
            print("\nAgenda del Martes:")
            if martes1 == "":
                print("Turno 1: (libre)")
            else:
                print(f"Turno 1: {martes1}")

            if martes2 == "":
                print("Turno 2: (libre)")
            else:
                print(f"Turno 2: {martes2}")

            if martes3 == "":
                print("Turno 3: (libre)")
            else:
                print(f"Turno 3: {martes3}")
            print()

    # ----------------------------------------
    # OPCIÓN 4: VER RESUMEN GENERAL
    # ----------------------------------------
    elif opcion == 4:
        # Contamos ocupados y libres en cada día SIN listas, sumando uno por uno

        # Lunes
        ocupados_lunes = 0
        if lunes1 != "":
            ocupados_lunes += 1
        if lunes2 != "":
            ocupados_lunes += 1
        if lunes3 != "":
            ocupados_lunes += 1
        if lunes4 != "":
            ocupados_lunes += 1

        libres_lunes = 4 - ocupados_lunes

        # Martes
        ocupados_martes = 0
        if martes1 != "":
            ocupados_martes += 1
        if martes2 != "":
            ocupados_martes += 1
        if martes3 != "":
            ocupados_martes += 1

        libres_martes = 3 - ocupados_martes

        print("\nResumen general:")
        print(f"Lunes -> Ocupados: {ocupados_lunes} | Libres: {libres_lunes}")
        print(f"Martes -> Ocupados: {ocupados_martes} | Libres: {libres_martes}")

        # Día con más turnos ocupados (o empate)
        if ocupados_lunes > ocupados_martes:
            print("Día con más turnos: Lunes\n")
        elif ocupados_martes > ocupados_lunes:
            print("Día con más turnos: Martes\n")
        else:
            print("Día con más turnos: Empate\n")

    # ----------------------------------------
    # OPCIÓN 5: CERRAR SISTEMA
    # ----------------------------------------
    elif opcion == 5:
        print("\nSistema cerrado. ¡Hasta luego!\n")
```

# Ejercicio 4 — “Escape Room: La Bóveda” 🏦🔐

## 📖 Historia
Sos un agente que intenta abrir una bóveda con **3 cerraduras**. Tenés **energía** y **tiempo** limitados.  
Si abrís las 3 cerraduras antes de quedarte sin energía o sin tiempo, ganás.

---

## ⚙️ Variables iniciales (NO se piden por teclado)
- `energia = 100`
- `tiempo = 12`
- `cerraduras_abiertas = 0`
- `alarma = False`
- `codigo_parcial = ""`

---

## ✅ Validaciones obligatorias
- ❌ No usar `try/except`
- Pedir **nombre del agente** y validar con `.isalpha()` dentro de un `while`
- Validar **opciones del menú** y cualquier número pedido con `.isdigit()` en `while`
- Usar secuenciales, condicionales y repetitivas (puede usar `.lower()`, `len()`, `f-strings`, etc.)

---

## 🚫 Regla anti-spam (muy importante)
Para evitar ganar forzando 3 veces seguidas al inicio:

Si el jugador elige **Forzar cerradura (opción 1) 3 veces seguidas**:
- se cobra el costo normal (**-20 energía, -2 tiempo**),
- **NO** abre cerradura,
- se activa la alarma automáticamente: `alarma = True` (“la cerradura se trabó”).

Si el jugador elige opción **2 o 3**, se corta la racha de forzar seguidas.

---

## 🎛️ Menú de acciones (se repite mientras el juego siga)
El juego continúa mientras:
- `energia > 0`, `tiempo > 0`, `cerraduras_abiertas < 3`
- y no esté bloqueado por alarma

### 1) Forzar cerradura (costo: -20 energía, -2 tiempo)
- Si `energia < 40`: hay “riesgo de alarma”
  - pedir un número **1-3** (validado)
  - si elige **3** → `alarma = True`
- Si no hay alarma, abre 1 cerradura
- Regla anti-spam: si es la 3ra vez seguida forzando, se activa alarma y no abre

### 2) Hackear panel (costo: -10 energía, -3 tiempo)
- Debe usar un `for` de **4 pasos** mostrando progreso
- En cada paso sumar una letra al `codigo_parcial` (por ejemplo `"A"`)
- Si `len(codigo_parcial) >= 8`, se abre automáticamente 1 cerradura si faltan

### 3) Descansar
- costo: `+15` energía (máx 100), `-1` tiempo
- si `alarma == True`: además `-10` energía extra

---

## 🚨 Regla de bloqueo por alarma
- Si `alarma == True` y `tiempo <= 3` y todavía no se abrió la bóveda → **bloqueo y derrota**

---

## 🏁 Condiciones de fin
- Si `cerraduras_abiertas == 3` → **VICTORIA**
- Si `energia <= 0` o `tiempo <= 0` → **DERROTA**
- Si hay bloqueo por alarma → **DERROTA (bloqueo)**


```python
# -----------------------------------------
# 0) VARIABLES INICIALES (no se piden)
# -----------------------------------------
# Estas variables representan el estado inicial del juego

energia = 100                  # Energía del agente (int)
tiempo = 12                    # Tiempo disponible (int)
cerraduras_abiertas = 0        # Cantidad de cerraduras abiertas (int)
alarma = False                 # Estado de la alarma (boolean)
codigo_parcial = ""            # Código acumulado durante el hackeo (string)

# Cuenta cuántas veces seguidas se eligió la opción 1 (forzar)
racha_forzar = 0

# Indica si el sistema quedó bloqueado por alarma + poco tiempo
bloqueo = False


# -----------------------------------------
# 1) PEDIR NOMBRE DEL AGENTE (solo letras)
# -----------------------------------------
agente = input("Nombre del agente: ").strip()

while agente == "" or not agente.isalpha():
    print("Error: ingresá un nombre no vacío y solo con letras.")
    agente = input("Nombre del agente: ").strip()

print(f"\nBienvenido/a, agente {agente}. Comienza la misión...\n")


# -----------------------------------------
# 2) BUCLE PRINCIPAL DEL JUEGO
# -----------------------------------------
# El juego se ejecuta mientras:
# - haya energía
# - haya tiempo
# - no se hayan abierto las 3 cerraduras
# - no exista bloqueo del sistema

while energia > 0 and tiempo > 0 and cerraduras_abiertas < 3 and bloqueo == False:

    # -----------------------------------------
    # 2.1 Verificar si hay bloqueo por alarma
    # -----------------------------------------
    # Si la alarma está activa y queda poco tiempo (<= 3),
    # el sistema se bloquea automáticamente.

    if alarma == True and tiempo <= 3 and cerraduras_abiertas < 3:
        bloqueo = True

    else:

        # -----------------------------------------
        # 2.2 Mostrar estado actual del juego
        # -----------------------------------------
        print("===== ESTADO =====")
        print(f"Energía: {energia}")
        print(f"Tiempo: {tiempo}")
        print(f"Cerraduras abiertas: {cerraduras_abiertas}/3")
        print(f"Alarma: {'ON' if alarma else 'OFF'}")
        print(f"Código parcial: '{codigo_parcial}' (len={len(codigo_parcial)})")
        print("==================")

        # -----------------------------------------
        # 2.3 Mostrar menú de acciones
        # -----------------------------------------
        print("1) Forzar cerradura (-20 energía, -2 tiempo)")
        print("2) Hackear panel     (-10 energía, -3 tiempo)")
        print("3) Descansar         (+15 energía máx 100, -1 tiempo; si alarma ON: -10 energía extra)")

        opcion_str = input("Elegí una opción (1-3): ").strip()

        # Validación estricta:
        # Debe ser número y estar dentro del rango 1-3
        while (not opcion_str.isdigit()) or (int(opcion_str) < 1) or (int(opcion_str) > 3):
            if not opcion_str.isdigit():
                print("Error: ingresá un número válido.")
            else:
                print("Error: opción fuera de rango (1 a 3).")
            opcion_str = input("Elegí una opción (1-3): ").strip()

        opcion = int(opcion_str)

        # ============================================================
        # USO DE MATCH (estructura alternativa a if / elif / else)
        # ============================================================
        # match evalúa el valor de "opcion"
        # y ejecuta el bloque cuyo case coincida.
        # Es una estructura de selección múltiple.

        match opcion:

            # -----------------------------------------
            # CASE 1 → FORZAR CERRADURA
            # -----------------------------------------
            case 1:

                # Se incrementa la racha de forzar seguidas
                racha_forzar += 1

                # Se aplican los costos de energía y tiempo
                energia -= 20
                tiempo -= 2

                # Si se forzó 3 veces seguidas, se activa alarma
                if racha_forzar == 3:
                    alarma = True
                    print("\n⚠ La cerradura se trabó por forzar 3 veces seguidas.")
                    print("¡Alarma activada! No se abrió ninguna cerradura.\n")

                else:

                    # Si la energía es baja, existe riesgo de activar alarma
                    if energia < 40:
                        print("\n⚠ Energía baja: hay riesgo de alarma.")
                        riesgo_str = input("Elegí un número (1-3): ").strip()

                        while (not riesgo_str.isdigit()) or (int(riesgo_str) < 1) or (int(riesgo_str) > 3):
                            print("Error: número inválido.")
                            riesgo_str = input("Elegí un número (1-3): ").strip()

                        riesgo = int(riesgo_str)

                        # Si elige 3, se activa la alarma
                        if riesgo == 3:
                            alarma = True
                            print("🚨 Se activó la alarma durante el intento.\n")

                    # Si no hay alarma, se abre una cerradura
                    if alarma == False and cerraduras_abiertas < 3:
                        cerraduras_abiertas += 1
                        print("✅ Cerradura abierta con éxito.\n")
                    else:
                        print("❌ No se pudo abrir la cerradura.\n")

            # -----------------------------------------
            # CASE 2 → HACKEAR PANEL
            # -----------------------------------------
            case 2:

                # Se corta la racha de forzar
                racha_forzar = 0

                # Se descuentan energía y tiempo
                energia -= 10
                tiempo -= 3

                print("\n💻 Iniciando hackeo...")

                # Bucle for que simula 4 pasos de hackeo
                # En cada paso se agrega una letra al código
                for paso in range(1, 5):
                    codigo_parcial += "A"
                    print(f"Paso {paso}/4... código parcial: '{codigo_parcial}'")

                # Si el código alcanza 8 caracteres, abre cerradura
                if len(codigo_parcial) >= 8 and cerraduras_abiertas < 3:
                    cerraduras_abiertas += 1
                    print("✅ Código suficiente. Cerradura abierta.\n")
                else:
                    print("ℹ Hackeo finalizado.\n")

            # -----------------------------------------
            # CASE 3 → DESCANSAR
            # -----------------------------------------
            case 3:

                # Se corta la racha de forzar
                racha_forzar = 0

                # Descansar consume tiempo
                tiempo -= 1

                # Recupera energía (máximo 100)
                energia += 15
                if energia > 100:
                    energia = 100

                # Si la alarma está activa, penaliza energía extra
                if alarma == True:
                    energia -= 10
                    print("\n😰 Descanso con alarma ON: -10 energía extra.")
                else:
                    print("\n😌 Descanso exitoso.")

                print(f"Energía actual: {energia}\n")


# -----------------------------------------
# 3) CONDICIONES DE FIN
# -----------------------------------------
# Se evalúan en orden lógico según prioridad

if cerraduras_abiertas == 3:
    print("🏆 VICTORIA: ¡Abriste las 3 cerraduras!")
elif bloqueo == True:
    print("🛑 DERROTA: sistema bloqueado por alarma.")
elif energia <= 0 or tiempo <= 0:
    print("💀 DERROTA: sin energía o sin tiempo.")
else:
    print("Fin del juego.")
```

# Ejercicio 5 — “Escape Room: La Arena del Gladiador” ⚔️🏟️

## 🧩 Descripción del escenario
Vas a desarrollar un simulador de batalla por turnos en Python.  
El usuario (Gladiador) pelea contra un enemigo controlado por la computadora.  
Gana quien reduzca primero la vida del rival a 0.

Este ejercicio evalúa:
- variables (`int`, `float`, `string`, `boolean`)
- estructuras de control (`if/elif/else`)
- ciclos (`while` y `for`)
- validación estricta de datos

---

## ✅ Requerimientos técnicos

### A) Tipos de datos obligatorios
- `string`: nombre del jugador
- `int`: HP (vida) y cantidad de pociones
- `float`: cálculo de daño crítico (multiplica por `1.5`)
- `boolean`: control del turno o si el juego sigue activo

### B) Validaciones obligatorias
- ❌ No usar `try/except`
- Validar texto con `.isalpha()` dentro de un `while`
- Validar números con `.isdigit()` dentro de un `while`

---

## 🔁 Flujo del programa

### Paso 1: Configuración
Pedir el nombre del Gladiador:
- Solo letras (no números, símbolos ni vacío)
- Si es inválido: mostrar **"Error: Solo se permiten letras."** y volver a pedir

### Paso 2: Inicialización (sin pedir al usuario)
Definir:
- Vida Gladiador: `100` (int)
- Vida Enemigo: `100` (int)
- Pociones: `3` (int)
- Daño base Ataque Pesado: `15` (int)
- Daño base enemigo: `12` (int)
- Turno Gladiador: `True` (booleano)

### Paso 3: Ciclo de combate
Repetir mientras ambos tengan vida > 0.

#### Turno del jugador
Mostrar HP de ambos y pociones. Menú:
1. Ataque Pesado
2. Ráfaga Veloz (usa `for`)
3. Curar

Validación del menú:
- debe ser número (`.isdigit()`)
- debe estar entre 1 y 3

Acciones:
- **1) Ataque Pesado**:  
  si vida enemigo < 20 → golpe crítico (`daño * 1.5`, float)
- **2) Ráfaga Veloz**:  
  `for` de 3 golpes, cada uno resta 5 y muestra mensaje
- **3) Curar**:
  - si pociones > 0: +30 vida y -1 poción
  - si pociones == 0: mostrar "¡No quedan pociones!" y perder el turno

#### Turno del enemigo
Después de la acción del jugador, el enemigo ataca:
- resta 12 a la vida del jugador
- muestra mensaje del ataque

### Paso 4: Fin
- si vida_jugador > 0 → **VICTORIA**
- si vida_jugador <= 0 → **DERROTA**


```python
print("--- BIENVENIDO A LA ARENA ---")

# -----------------------------------------
# PASO 1: PEDIR NOMBRE (string) y VALIDAR
# -----------------------------------------
nombre = input("Nombre del Gladiador: ").strip()

# Validación:
# - No vacío
# - Solo letras
while nombre == "" or not nombre.isalpha():
    print("Error: Solo se permiten letras.")
    nombre = input("Nombre del Gladiador: ").strip()


# -----------------------------------------
# PASO 2: ESTADÍSTICAS INICIALES (int / bool)
# -----------------------------------------
vida_jugador = 100          # int
vida_enemigo = 100          # int
pociones = 3                # int
dano_ataque_pesado = 15     # int
dano_enemigo = 12           # int
turno_gladiador = True      # boolean
print("=== INICIO DEL COMBATE ===")


# -----------------------------------------
# PASO 3: CICLO PRINCIPAL DE COMBATE (while)
# -----------------------------------------
# El combate continúa mientras ambos tengan vida mayor a 0

while vida_jugador > 0 and vida_enemigo > 0:

    # -----------------------------
    # TURNO DEL JUGADOR
    # -----------------------------
    print(f"{nombre} (HP: {vida_jugador}) vs Enemigo (HP: {vida_enemigo}) | Pociones: {pociones}")
    print("Elige acción:")
    print("1. Ataque Pesado")
    print("2. Ráfaga Veloz")
    print("3. Curar")

    opcion_str = input("Opción: ").strip()

    # Validación estricta del menú:
    # - Debe ser número
    # - Debe estar entre 1 y 3
    while (not opcion_str.isdigit()) or (int(opcion_str) < 1) or (int(opcion_str) > 3):
        if not opcion_str.isdigit():
            print("Error: Ingrese un número válido.")
        else:
            print("Error: Opción fuera de rango (1 a 3).")
        opcion_str = input("Opción: ").strip()

    opcion = int(opcion_str)

    # =============
    # USO DE MATCH
    # =============
    # match evalúa el valor de "opcion"
    # y ejecuta el bloque correspondiente al case que coincida.

    match opcion:

        # -----------------------------------------
        # CASE 1 → ATAQUE PESADO
        # -----------------------------------------
        case 1:

            # Daño base (int)
            dano_final = dano_ataque_pesado

            # Si el enemigo tiene menos de 20 HP,
            # se aplica multiplicador 1.5 (float)
            if vida_enemigo < 20:
                dano_final = dano_ataque_pesado * 1.5
                print(">> ¡Golpe Crítico!")

            # Se descuenta el daño al enemigo
            vida_enemigo -= dano_final

            # Mostramos daño formateado a 1 decimal
            print(f"¡Atacaste al enemigo por {dano_final:.1f} puntos de daño!")


        # -----------------------------------------
        # CASE 2 → RÁFAGA VELOZ (usa FOR)
        # -----------------------------------------
        case 2:

            print(">> ¡Inicias una ráfaga de golpes!")

            # Bucle for que ejecuta 3 golpes
            for _ in range(3):
                vida_enemigo -= 5
                print("> Golpe conectado por 5 de daño")


        # -----------------------------------------
        # CASE 3 → CURAR
        # -----------------------------------------
        case 3:

            # Solo se puede curar si hay pociones disponibles
            if pociones > 0:
                vida_jugador += 30
                pociones -= 1
                print(">> Usaste una poción y recuperaste 30 HP.")
            else:
                print("¡No quedan pociones!")
                # Aunque no cure, el enemigo igual atacará


    # -----------------------------
    # TURNO DEL ENEMIGO (AUTOMÁTICO)
    # -----------------------------
    # El enemigo siempre ataca después del turno del jugador,
    # independientemente de la acción elegida.

    vida_jugador -= dano_enemigo
    print(f">> ¡El enemigo contraataca por {dano_enemigo} puntos!")

    print("=== NUEVO TURNO ===")


# -----------------------------------------
# PASO 4: FIN DEL JUEGO
# -----------------------------------------
if vida_jugador > 0:
    print(f"¡VICTORIA! {nombre} ha ganado la batalla.")
else:
    print("DERROTA. Has caído en combate.")
```
