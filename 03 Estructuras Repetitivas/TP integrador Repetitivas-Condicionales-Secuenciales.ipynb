```python
# 1) Kiosko

while True:
    nombre = input("Ingrese nombre del cliente : ")
    if nombre.isalpha():
        break
    else:
        print("Error: el nombre debe contener solo letras.")

while True:
    cantidad = input("Ingrese cantidad de productos : ")
    if cantidad.isdigit() and int(cantidad) > 0:
        cantidad = int(cantidad)
        break
    else:
        print("Error : ingrese un número mayor a 0.")

total_sin_descuento = 0
total_con_descuento = 0

productos = []

for i in range(1, cantidad + 1):
    while True:
        precio = input(f"Producto {i} - Precio: ")
        if precio.isdigit():
            precio = int(precio)
            break
        else:
            print("Error : ingrese un número válido.")

    while True:
        descuento = input("Descuento (S/N): ").lower()
        if descuento in ["s", "n"]:
            break
        else:
            print("Error : ingrese S o N.")

    total_sin_descuento += precio

    if descuento == "s":
        precio_final = precio * 0.9
    else:
        precio_final = precio

    total_con_descuento += precio_final

    productos.append((i, precio, descuento))

ahorro = total_sin_descuento - total_con_descuento
promedio = total_con_descuento / cantidad

print(f"Cliente: {nombre}")
print(f"Cantidad de productos : {cantidad}")
for p in productos:
    print(f"Producto {p[0]} - Precio : {p[1]}  Descuento (S/N): {p[2]}")
print(f"Total sin descuentos : ${total_sin_descuento}")
print(f"Total con descuentos: ${total_con_descuento:.2f}")
print(f"Ahorro : ${ahorro:.2f}")
print(f"Promedio por producto : ${promedio:.2f}")
```

```python
# 3) “Agenda de Turnos con Nombres (sin listas)

def solo_letras(texto):
    return texto.isalpha()


# Inicialización de turnos
lunes = [""] * 4
martes = [""] * 3


def elegir_dia():
    while True:
        dia = input("Elegir día (1=Lunes, 2=Martes): ")
        if dia == "1":
            return lunes, "Lunes"
        elif dia == "2":
            return martes, "Martes"
        else:
            print("Opción inválida.")


def reservar():
    agenda, nombre_dia = elegir_dia()

    paciente = input("Nombre del paciente: ")
    if not solo_letras(paciente):
        print("Nombre inválido (solo letras).")
        return

    # Verificar repetido
    if paciente in agenda:
        print("Ese paciente ya tiene turno ese día.")
        return

    # Buscar espacio libre
    for i in range(len(agenda)):
        if agenda[i] == "":
            agenda[i] = paciente
            print(f"Turno reservado en {nombre_dia}, turno {i+1}.")
            return

    print("No hay turnos disponibles.")


def cancelar():
    agenda, nombre_dia = elegir_dia()

    paciente = input("Nombre del paciente: ")
    if not solo_letras(paciente):
        print("Nombre inválido.")
        return

    if paciente in agenda:
        index = agenda.index(paciente)
        agenda[index] = ""
        print(f"Turno cancelado en {nombre_dia}.")
    else:
        print("Paciente no encontrado.")


def ver_agenda():
    agenda, nombre_dia = elegir_dia()

    print(f"\nAgenda de {nombre_dia}:")
    for i, turno in enumerate(agenda):
        if turno == "":
            print(f"Turno {i+1}: (libre)")
        else:
            print(f"Turno {i+1}: {turno}")


def resumen():
    ocupados_lunes = sum(1 for t in lunes if t != "")
    ocupados_martes = sum(1 for t in martes if t != "")

    libres_lunes = len(lunes) - ocupados_lunes
    libres_martes = len(martes) - ocupados_martes

    print("\nResumen:")
    print(f"Lunes -> Ocupados: {ocupados_lunes}, Libres: {libres_lunes}")
    print(f"Martes -> Ocupados: {ocupados_martes}, Libres: {libres_martes}")

    if ocupados_lunes > ocupados_martes:
        print("Día con más turnos: Lunes")
    elif ocupados_martes > ocupados_lunes:
        print("Día con más turnos: Martes")
    else:
        print("Hay empate entre ambos días.")


# MAIN
operador = input("Nombre del operador: ")
while not solo_letras(operador):
    operador = input("Nombre inválido. Ingrese solo letras: ")

print(f"Bienvenido {operador}!")

while True:
    print("\n--- MENÚ ---")
    print("1. Reservar turno")
    print("2. Cancelar turno")
    print("3. Ver agenda del día")
    print("4. Ver resumen general")
    print("5. Cerrar sistema")

    opcion = input("Elegir opción: ")

    if opcion == "1":
        reservar()
    elif opcion == "2":
        cancelar()
    elif opcion == "3":
        ver_agenda()
    elif opcion == "4":
        resumen()
    elif opcion == "5":
        print("Sistema cerrado.")
        break
    else:
        print("Opción inválida.")
```

```python
# 4) “Escape Room: La Bóveda"

energia = 100
tiempo = 12
cerraduras_abiertas = 0
alarma = False
codigo_parcial = ""

racha_forzar = 0

agente = input("Nombre del agente: ")
while not agente.isalpha():
    agente = input("Nombre inválido. Solo letras: ")

print(f"\nBienvenido agente {agente}!")

while energia > 0 and tiempo > 0 and cerraduras_abiertas < 3:
    if alarma and tiempo <= 3:
        print("\n🚨 La alarma bloqueó la bóveda. PERDISTE.")
        break

    print("\n--- ESTADO ---")
    print(f"Energía: {energia}")
    print(f"Tiempo: {tiempo}")
    print(f"Cerraduras abiertas: {cerraduras_abiertas}")
    print(f"Alarma: {'ON' if alarma else 'OFF'}")

    print("\n1. Forzar cerradura (-20 energía, -2 tiempo)")
    print("2. Hackear panel (-10 energía, -3 tiempo)")
    print("3. Descansar (+15 energía, -1 tiempo)")

    opcion = input("Elegir opción: ")
    while not opcion.isdigit() or opcion not in ["1", "2", "3"]:
        opcion = input("Opción inválida. Elegir 1, 2 o 3: ")

    # OPCIÓN 1: FORZAR
    if opcion == "1":
        energia -= 20
        tiempo -= 2
        racha_forzar += 1
        if racha_forzar == 3:
            print(" Forzaste 3 veces seguidas. La cerradura se trabó!")
            alarma = True
            continue

        if energia < 40:
            riesgo = input("Riesgo de alarma! Elegir número (1-3): ")
            while not riesgo.isdigit() or riesgo not in ["1", "2", "3"]:
                riesgo = input("Número inválido. Elegir 1, 2 o 3: ")

            if riesgo == "3":
                print("🚨 Se activó la alarma!")
                alarma = True

        if not alarma:
            cerraduras_abiertas += 1
            print(" Abriste una cerradura!")

    # OPCIÓN 2: HACKEAR
    elif opcion == "2":
        energia -= 10
        tiempo -= 3
        racha_forzar = 0 

        print("Hackeando panel...")
        for i in range(4):
            print(f"Progreso {i+1}/4")
            codigo_parcial += "A"

        print(f"Código parcial: {codigo_parcial}")

        if len(codigo_parcial) >= 8 and cerraduras_abiertas < 3:
            cerraduras_abiertas += 1
            print(" Código completo! Se abrió una cerradura.")

    # OPCIÓN 3: DESCANSAR
    elif opcion == "3":
        energia += 15
        if energia > 100:
            energia = 100

        tiempo -= 1
        racha_forzar = 0  

        if alarma:
            energia -= 10
            print(" La alarma consume energía extra!")

        print(" Descansaste.")

# RESULTADO
if cerraduras_abiertas == 3:
    print("\n VICTORIA! Abriste la bóveda.")
elif energia <= 0 or tiempo <= 0:
    print("\n DERROTA! Te quedaste sin recursos.")
elif alarma and tiempo <= 3:
    print("\n DERROTA! Sistema bloqueado por alarma.")
```

```python
# 5) “Escape Room:"La Arena del Gladiador"

# Nombre
nombre = input("Nombre del Gladiador: ")
while not nombre.isalpha():
    print("Error: Solo se permiten letras.")
    nombre = input("Nombre del Gladiador: ")

# Variables
vida_jugador = 100
vida_enemigo = 100
pociones = 3
daño_pesado = 15
daño_enemigo = 12
turno_jugador = True

juego_activo = True

print("=== INICIO DEL COMBATE ===")

while vida_jugador > 0 and vida_enemigo > 0:

    print(f"\n{nombre} (HP: {vida_jugador}) vs Enemigo (HP: {vida_enemigo}) | Pociones: {pociones}")
    print("Elige acción:")
    print("1. Ataque Pesado")
    print("2. Ráfaga Veloz")
    print("3. Curar")

    opcion = input("Opción: ")
    while not opcion.isdigit() or opcion not in ["1", "2", "3"]:
        print("Error: Ingrese un número válido.")
        opcion = input("Opción: ")

    # Ataque Pesado
    if opcion == "1":
        daño = daño_pesado
        if vida_enemigo < 20:
            daño = daño * 1.5

        vida_enemigo -= daño
        print(f"Atacaste al enemigo por {daño} puntos de daño")

    # Ráfaga Veloz
    elif opcion == "2":
        print("Inicias una ráfaga de golpes")
        for i in range(3):
            vida_enemigo -= 5
            print("> Golpe conectado por 5 de daño")

    # Curar
    elif opcion == "3":
        if pociones > 0:
            vida_jugador += 30
            pociones -= 1
            print("Recuperaste 30 de vida")
        else:
            print("No quedan pociones")

    # Turno enemigo (si sigue vivo)
    if vida_enemigo > 0:
        vida_jugador -= daño_enemigo
        print(f"El enemigo te atacó por {daño_enemigo} puntos de daño")

# Resultado
if vida_jugador > 0:
    print(f"VICTORIA! {nombre} ha ganado la batalla.")
else:
    print("DERROTA. Has caído en combate.")
```
